```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the dataset
df = pd.read_csv('sales_data.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
# Group by product_id and calculate total revenue for each product
product_revenue = df.groupby('product_id')['revenue'].sum()
```

```python
# Sort products by revenue in descending order and select top 10 products
top_products = product_revenue.sort_values(ascending=False).head(10)
```

```python
# Create a bar plot to visualize the top 10 products by revenue
plt.figure(figsize=(10, 6))
top_products.plot(kind='bar', color='blue')
plt.title('Top 10 Products by Revenue')
plt.xlabel('Product ID')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
