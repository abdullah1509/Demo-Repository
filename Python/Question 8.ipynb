```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Fetch the data from the API
response = requests.get('http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes')
data = response.json()
```

```python
# Extract the episode data
episodes = data['_embedded']['episodes']

# Create a DataFrame from the episode data
df = pd.DataFrame(episodes)
```

```python
# Analyze the data and draw plots
# Example: Plot the ratings of each episode
plt.figure(figsize=(10, 6))
plt.plot(df['season'], df['number'], marker='o', linestyle='-', color='b')
plt.xlabel('Season')
plt.ylabel('Episode Number')
plt.title('Episode Ratings')
plt.grid(True)
plt.show()
```
